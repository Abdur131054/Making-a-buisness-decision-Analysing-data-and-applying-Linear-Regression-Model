## Importing necessary Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

## Get the Data

We'll work with the Ecommerce Customers csv file from the company. It has Customer info, suchas Email, Address, and their color Avatar. Then it also has numerical value columns:

* Avg. Session Length: Average session of in-store style advice sessions.
* Time on App: Average time spent on App in minutes
* Time on Website: Average time spent on Website in minutes
* Length of Membership: How many years the customer has been a member. 

** Read in the Ecommerce Customers csv file as a DataFrame called customers.**

```python
#read a file
customer1=pd.read_excel(r'Ecommerce Customers.xlsx',sheet_name='Ecommerce Customer 1')
```

```python
#Read another file
customer2=pd.read_excel(r'Ecommerce Customers.xlsx',sheet_name='Ecommerce Customer 2')
```

# Join and Clean the data

```python
#Concat/join two file those were imported
customer=pd.concat([customer1, customer2],ignore_index=True)
```

```python
#check column's names of the data frame
customer.columns
```

Checking and filling Missing Values

```python
#Checking missing values quantity
customer.isna().sum()
```

```python
#Checking null values in percentages  
customer.isnull().mean()*100
```

We have  missing values on 'Avg. Session Length', 'Time on Website' and 'Length of Membership ' columns. To fill the missing value we need to check mean, median,std ,range , box plot and distribution plot

```python
#Checking mean, median, mode ,std , min and max of 'Avg. Session Length'
print(customer['Avg. Session Length'].mean())
print(customer['Avg. Session Length'].median())
print(customer['Avg. Session Length'].mode()[0])
print(customer['Avg. Session Length'].std())
print(customer['Avg. Session Length'].min())
print(customer['Avg. Session Length'].max())
```

```python
sns.boxplot(customer['Avg. Session Length'])
plt.show()
```

```python
sns.distplot(customer['Avg. Session Length'])
plt.show()
```

so from above information we observe that difference between values are not big and mean and Median values almost same. So we can use mean value to replace.
And also we observe from the plot that there dont have much outlier and data not skewed much. so using Mean value wont create any problem 

And from plot we can say that it is almost symetrical distribution , and for symetrical distribution we can use mean value to impute missing or null Value

```python
#checking mean value of 'Avg. Session Length' and pass it to a variable nameed 'a'
a= customer['Avg. Session Length'].mean()
```

```python
#Filling null values of 'Avg. Session Length' with it's mean values
customer['Avg. Session Length'].fillna(a, inplace=True)
```

```python
#Checking mean, median , mode, std, min and max values of 'Time on Website'
print(customer['Time on Website'].mean())
print(customer['Time on Website'].median())
print(customer['Time on Website'].mode()[0])
print(customer['Time on Website'].std())
print(customer['Time on Website'].min())
print(customer['Time on Website'].max())
```

```python
sns.boxplot(customer['Time on Website'])
```

```python
sns.distplot(customer['Time on Website'])
```

Here also mean and median value almost same.And there have not much outliers. So, We are going to impute missing value with mean value .

```python
#checking mean value of 'Time on Website' and pass it to a variable named 'b'
b=customer['Time on Website'].mean()
```

```python
#filling missing values of 'Time on Website' column's with it's mean value
customer['Time on Website'].fillna(b, inplace=True)
```

```python
#Checking mean, median, mode, std , min and max values of 'Length of Membership'
print(customer['Length of Membership'].mean())
print(customer['Length of Membership'].median())
print(customer['Length of Membership'].mode()[0])
print(customer['Length of Membership'].std())
print(customer['Length of Membership'].min())
print(customer['Length of Membership'].max())
```

```python
#draw boxplot for 'Length of Membership' columns 
sns.boxplot(customer['Length of Membership'])
```

```python
#plot distbriution of 'Length of Membership'
sns.displot(customer['Length of Membership'])
plt.show()
```

Although mean and median values almost same and data distribution is symmatric, so in this logic we can use mean value for imputation. but from boxplot we observe that there have some outliers it is not recommended to use mean value. so we will use median values for imputation

```python
#check median value of 'Length of Membership' and pass it to a variable named 'c'
c=customer['Length of Membership'].median()
```

```python
#filling missing values of 'Length of Membership'
customer['Length of Membership'].fillna(c, inplace=True)
```

## Training and Testing Data

Now that we've explored the data a bit, let's go ahead and split the data into training and testing sets.
** Set a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column. **

```python
#Check columns name
customer.columns
```

```python
#all numerical features of customer without 'Yearly amount Spent'
X=customer[[ 'Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
#set target variable
y=customer['Yearly Amount Spent']
```

```python
# importing train_test_split 
from sklearn.model_selection import train_test_split
```

```python
# perform train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3, random_state=101)
```

## Training the Model

Now its time to train our model on our training data!

** Import LinearRegression from sklearn.linear_model **

```python
from sklearn.linear_model import LinearRegression
```

**Create an instance of a LinearRegression() model named lm.**

```python
#creating instance for LinearRegression
lm = LinearRegression()
```

** Train/fit lm on the training data.**

```python
#fit the model on training data
lm.fit(X_train,y_train)
```

**Print out the coefficients of the model**

```python
 pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
#coeff_df
```

## Predicting Test Data
Now that we have fit our model, let's evaluate its performance by predicting off the test values!

** Use lm.predict() to predict off the X_test set of the data.**

```python
#predicting X_test set of data
predictions = lm.predict(X_test)
```

** Create a scatterplot of the real test values versus the predicted values. **

```python
plt.scatter(y_test,predictions)
plt.show()
```

## Evaluating the Model

Let's evaluate our model performance by calculating the residual sum of squares and the explained variance score (R^2).

** Calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error. **

```python
#import metrics 
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('Median absolute error:', metrics.median_absolute_error(y_test, predictions))
print('Mean squared log error:', metrics.mean_squared_log_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print('R2:', metrics.r2_score(y_test, predictions))
```

## Residuals

 Let's quickly explore the residuals to make sure everything was okay with our data. 

**Plot a histogram of the residuals and make sure it looks normally distributed. 

```python
sns.displot(y_test-predictions,bins=50);
```
